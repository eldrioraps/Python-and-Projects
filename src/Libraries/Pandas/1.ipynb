```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
chipo = pd.read_csv(url, sep = '\t')
```

```python
c = chipo.groupby('item_name').agg({
    'quantity': 'sum',       
    'order_id': 'count',
    'item_price':'sum'     
})
c.rename(columns={'order_id': 'order_count'}, inplace=True)
c = c.sort_values('quantity', ascending=False)
c.head(5)
```

```python
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
item_sales = chipo.groupby('item_name')['revenue'].sum().reset_index()

item_sales['rank'] = item_sales['revenue'].rank(ascending=False, method='dense')
```
